```python
Task-1:
Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show
x – Time taken in seconds
y – Distance reached
Title “Buzzer Game Results”
One you have a working program, add the following code to your program;
colors=["red","green","blue","yellow","pink","orange","purple","brown"
,"cyan","magenta"]
And add to your scatter() call the ‘c=colors’ argument like so;
plt.scatter(x, y, c=colors)
What happens?
```

```python
import matplotlib.pyplot as plt
import random

x = [random.randint(0, 51) for i in range(10)]
y = [random.randint(0, 51) for i in range(10)]

colors = ["red", "green", "blue", "yellow", "pink", "orange", "purple", "brown", "cyan", "magenta"]

plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")
plt.scatter(x, y, c=colors)
plt.show()
```

```python
Task 2:
Following data shows the change in population of different animals from 2017 to 2022.
a. Create a pie chart for the population of each year
b. Can we show all the pie chart on the single pane? Write the code.

     Animal Populations by Year
                     
Year	Bears	Dolphins	Whales
2017	8	      150	      80
2018	54	      77	      54
2019	93	      32	      100
2020	116    	  11	      76
2021	137  	  6	          93
2022	184	      1 	      72
```

```python
import matplotlib.pyplot as plt 
populationByYear = {
    2017: {'Bears': 8, 'Dolphins': 150, 'Whales': 80},
    2018: {'Bears': 54, 'Dolphins': 77, 'Whales': 54},
    2019: {'Bears': 93, 'Dolphins': 32, 'Whales': 100},
    2020: {'Bears': 116, 'Dolphins': 11, 'Whales': 76},
    2021: {'Bears': 137, 'Dolphins': 6, 'Whales': 93},
    2022: {'Bears': 184, 'Dolphins': 1, 'Whales': 72}
}
plt.figure(figsize=(15, 10))
plt.suptitle('Pie chart representing the population of animals from 2017 - 2022')
count = 1
for year in populationByYear.keys():
    plt.subplot(3, 3, count)
    data = populationByYear[year]
    animal = list(data.keys())
    population = list(data.values())
    plt.pie(population, labels=animal)
    plt.title(f'Animal Population in {year}')
    count += 1

plt.tight_layout(pad = 4)
plt.show()
```

```python
Task 3:
Construct a pie chart to visually display the favorite fruits of the students in a class based on the
given data: Mango - 45; Orange - 30; Plum - 15; Pineapple - 30; Melon – 30
```

```python
import matplotlib.pyplot as plt

favFruit = {'Mango': 45, 'Orange': 30, 'Plum': 15, 'Pineapple': 30, 'Melon': 30}
plt.pie(list(favFruit.values()), labels=list(favFruit.keys()))
plt.legend(bbox_to_anchor=(0, 1))
plt.title('Favorite Fruits of Students')
plt.show()
```
